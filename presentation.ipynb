# Problem 

People with disabilities may face very different challenges with health and in accessing health care than people without. 

Through this project, I'm hoping to explore which factors most impact health for people with and without disabilities, including states' Medicaid spending and quality of care. 


# Background


I'm using data from the CDC's Behavior Risk Factor Surveillance Survey. It's administered yearly overy the phone, and includes information on health conditions, behaviors, and access to care, along with demographic features for respondents. 

I am also using data about Medicaid provided on Medicaid.gov. So far, this includes quality of care ratings for each state's Medicaid recipients. I plan to include more information about enrollment and spending moving forward in my analysis. 

A significant limitation in this work is the lack of available data on people living in certain group or institutional settings. Long term residential services are a significant portion of Medicaid spending, and the experiences of people living within these settings can vary widely from people living within their communities. 


# Methods

To begin, I've been exploring different factors which may influence health outcomes, using BRFSS survey respodents' self-rating of health as my dependent variable. 

I'm starting by including a wide range of factors, including demographic variables such as race, gender, sexual orientation, and employment status; and health factors including amount of exercise, certain medical diagnoses, and ability to access medical care. 

I ran my dataset through the pipeline we used in previous class projects to start identifying which variables have the greatest impact.

Although my work so far has focused on the individual level to determine which factors are most important in health outcomes,I have also started to look at my data by state.
